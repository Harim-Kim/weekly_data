```python
csv_all_column["sequece"] = csv_all_column["Issue key"].str[5:].astype(int)
csv_all_column = csv_all_column.sort_values(by='sequece')
csv_all_column.index = np.arange(1, len(csv_all_column)+1)
```

```python
csv_all_column = csv_all_column.rename(columns={'Security Level':'Region','Issue Type':'Type', 'Issue key':'ID',"Summary":"Descriptions","Created":"Reported Date","Updated":"Resolved date" })
```

```python
csv_report = csv_all_column[['Region','Type',"ID","Descriptions","Reporter","Status","Reported Date","Resolved date"]]
```

```python
status = {'Closed':'Done',
          "Done":"Done",
          'Backlog':'Backlog',
          'Customer Review':'Awaition confrimation',
          'Clarification Required':'In review',
          'HAE in progress':'In development(HAE)',
          'In Progress (HAEE)':'In development(HAEE)',
          'New':'In review',
          'Under Review':'In review'
          }
csv_report = csv_report.replace({"Status":status})
```

```python
csv_report
```

```python
csv_report["Reported Date"] = csv_report["Reported Date"].str[:10]
csv_report["Reported Date"] = pd.to_datetime(csv_report["Reported Date"], format='%Y-%m-%d')
csv_report.loc[csv_report.Status != 'Done', 'Resolved date'] = ''
csv_report["Resolved date"] = pd.to_datetime(csv_report["Resolved date"], format='%Y-%m-%d')
```

```python
csv_report["Reported Date"] = csv_report["Reported Date"].dt.strftime('%d-%b-%Y')

csv_report["Resolved date"] = csv_report["Resolved date"].dt.strftime('%d-%b-%Y')
```
