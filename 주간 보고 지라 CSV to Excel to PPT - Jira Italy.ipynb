```python
csv_all["Reported Date"] = csv_all["Reported Date"].str[:10]
csv_all["Reported Date"] = pd.to_datetime(csv_all["Reported Date"], format='%Y-%m-%d')
csv_all.loc[csv_all.Status != 'Done', 'Resolved date'] = ''
csv_all["Resolved date"] = pd.to_datetime(csv_all["Resolved date"], format='%Y-%m-%d')
```

```python
csv_all["Reported Date"] = csv_all["Reported Date"].dt.strftime('%d-%b-%Y')

csv_all["Resolved date"] = csv_all["Resolved date"].dt.strftime('%d-%b-%Y')
```
